# Data Staging - Railway Dataset

## 1. Import Libraries

```python
import pandas as pd
from sqlalchemy import create_engine

print("Libraries Imported Successfully")
```

## 2. Setup Database Connection (SQL Server)

```python
server = r".\SQLEXPRESS"
staging_database = "staging_railway"


connection_string = (
    f"mssql+pyodbc://{server}/{staging_database}"
    "?driver=ODBC+Driver+18+for+SQL+Server"
    "&TrustServerCertificate=yes"
)

engine = create_engine(connection_string)
print("Engine created successfully!")
```

## 3. Load Raw Data and Push to Staging

```python
df_railway = pd.read_csv(r"D:\Dashboards\Railway_Project\data\raw\railway.csv")

df_railway.to_sql("railway", engine, if_exists="replace", index=False)

print("Railway data loaded into staging table 'railway'!")
```
